<h2 style='color: purple'>Initialize and load dataset</h2>

```python
airline_passenger_satisfaction = pd.read_csv('dataset/airline_passenger_satisfaction.csv', sep=';')
```

<h2 style='color: purple;'>Check missing value</h2>

<h3 style='color: orange'>Check each feature for missing values</h3>

Jumlah data yang missing sebanyak 393 atau 0,302 persen dari jumlah keselurahan data pada fitur Arrival Delay in Minutes

```python
airline_passenger_satisfaction.isnull().sum()
```

<h3 style='color: orange'>Delete the line that is missing value</h3>
