<h1 style='color: green; text-align:center'>Visualization for each Feature: Passenger Satisfaction Dataset - US Airline Passenger Satisfaction</h1>

<h2 style='color: purple;'>Import libraries and packages</h2>

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set()
```

<h2 style='color: purple'>Initialize and load dataset</h2>

```python
airline_passenger_satisfaction_no_mv = pd.read_csv('dataset/airline_passenger_satisfaction_no_mv.csv')
```

```python
pd.set_option('display.max_columns', None)
airline_passenger_satisfaction_no_mv.head()
```

<h2 style='color: purple'>Heatmap Plot</h2>

```python
airline_passenger_satisfaction_replace = airline_passenger_satisfaction_no_mv.copy()

airline_passenger_satisfaction_replace['Satisfaction']   = airline_passenger_satisfaction_no_mv['Satisfaction'].map({'neutral or dissatisfied':0, 'satisfied':1})
airline_passenger_satisfaction_replace['Gender']         = airline_passenger_satisfaction_no_mv['Gender'].map({'Female':0, 'Male':1})
airline_passenger_satisfaction_replace['Customer Type']  = airline_passenger_satisfaction_no_mv['Customer Type'].map({'disloyal Customer':0, 'Loyal Customer':1})
airline_passenger_satisfaction_replace['Type of Travel'] = airline_passenger_satisfaction_no_mv['Type of Travel'].map({'Business travel':0, 'Personal Travel':1})
airline_passenger_satisfaction_replace['Class']          = airline_passenger_satisfaction_no_mv['Class'].map({'Eco':0, 'Eco Plus':1, 'Business':2})
```

```python
airline_passenger_satisfaction_replace.nunique()
```

<h3 style='color: orange'>Keterangan</h3>

- <b>Satisfaction</b> = 0:neutral or satisfied and 1:satisfied 
- <b>Gender</b> = 0: Female and 1: Male
- <b>Customer Type</b> = 0: disloyal Customer and 1:Loyal Customer
- <b>Type of Travel</b> = 0:Business travel and 1:Personal Travel
- <b>Class</b> = 0:Eco; 1:Eco Plus; and 2:Business

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
airline_passenger_satisfaction_replace.head()
```

```python
plt.figure(figsize=(20,12))

c = airline_passenger_satisfaction_replace.corr()
sns.heatmap(c, annot=True)
```

<h3 style='color: orange'>Pola dan hubungan menarik antar variable</h3>

- Flight Distance dan Age, semakin jauh jarak penerbangan, maka umur dari penumpang pesawat akan semakin rendah.
- Food and Drink dan Seat comfort cukup memiliki hubungan yang kuat.
- Delay keberangkatan dan kedatangan tidak memperngaruhi fitur-fitur yang ada.
- Umur tidak berpengaruh terhadap nilai rating yang diberikan pada tiap fitur.
- Departure Delay in Minutes dan Arrival Delay in Minutes hubungannya linier.
